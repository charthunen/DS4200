### Part 3 - Basic data analysis with Altair (40 points)

We are going to use the same Weather data from Homework 1 but use Altair to create the data visualziation. 

##### Question 1 (10 points)

Recreate the figure in Part 3 Question 2 in Homework 1 with Altair. Include necessary title and labels. Here is the question again:
    
    Generate a data visualization to show the relationship between max temperature and weather type.

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import altair as alt
from vega_datasets import data

weather = data.seattle_weather()
weather.head()
```

```python
# Create box plot to show the relationship between max temp and weather type
max_weather = alt.Chart(weather).mark_boxplot().encode(
    x=alt.X('weather:N', title='Weather Type'),
    y=alt.Y('temp_max:Q', title='Max Temperature (C)'),
    color='weather:N'
).properties(
    title="Relationship Between Max Temperature and Weather Type",
    width=400,
    height=325
)

max_weather
```

##### Question 2 (10 points)

Recreate the figure in Part 3 Question 3 in Homework 1 with Altair. Include necessary title and labels. Here is the question again:
    
    Generate a data visualization to show the relationship between max temperature and min temperature, and then show different color for differnet weather type.

```python
min_max = alt.Chart(weather).mark_circle(size=60, opacity=0.7).encode(
    x=alt.X('temp_min:Q', title='Min Temperature (C)'),
    y=alt.Y('temp_max:Q', title='Max Temperature (C)'),
    color=alt.Color('weather:N', title='Weather Type'),
).properties(
    title="Max Temperature vs. Min Temperature by Weather Type",
    width=500,
    height=400
)

min_max
```

##### Question 3 (10 points)

Create a facet figure in Altair with 2 row and 3 columns. For each panel in the facet figure, it shows a scatter plot between the max and min temperature for one type of weather. Use the wind speed to color the points. Also add interaction to show the max and min temperature and wind speed for each data point. Include necessary title and labels. 

```python
# Create Facet Figure
facet_fig = alt.Chart(weather).mark_circle(size=60, opacity=0.7).encode(
    x=alt.X('temp_min:Q', title='Min Temperature (C)'),
    y=alt.Y('temp_max:Q', title='Max Temperature (C)'),
    color=alt.Color('wind:Q', title='Wind Speed (m/s)'),
    tooltip=['temp_min', 'temp_max', 'wind'] # add for interaction 
).add_params(
    alt.selection_point(on='mouseover', fields=['temp_min', 'temp_max', 'wind']) # interactive features
).properties(
    width=200,
    height=200
).facet(
    facet=alt.Facet('weather:N', title='Weather Type'),
    columns=3
).interactive()

facet_fig
```

```python
# Calculate the average temp
weather['temp_avg'] = (weather['temp_min'] + weather['temp_max']) / 2

# Create figure
base_fig = alt.Chart(weather).encode(
    x=alt.X('date:T', title='Date'),
    y=alt.Y('temp_avg:Q', title='Average Temperature (C)')
)

# Create marks - line and points
# Line for the trend of average temperature
line = base_fig.mark_line(color='blue', opacity=0.7).encode()

# Points for individual data points
points = base_fig.mark_circle(color='red', size=30, opacity=0.5).encode(
    tooltip=['date', 'temp_avg', 'temp_min', 'temp_max']  # Display date and temps on hover
)

# Create compound figure with the marks
comp_fig = (line + points).properties(
    title="Trend of Average Temperature Over 4 Years"
).interactive()  # Make figure interactive to allow zooming in/out

compound_chart
```
